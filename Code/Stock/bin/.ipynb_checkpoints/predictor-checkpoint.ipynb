```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
Stocks= ["ITC.NS", "BA", "DAI.DE", "EXPE", "GE", "HOG", "MMM", "XOM"]
```

```python
for S in Stocks:
    Pkl_Filename = ".//model_data//"+S+"_Model.pkl"
    sc_file= ".//model_data//"+S+"_sc.pkl"
    global sc
    with open(Pkl_Filename, 'rb') as file:  
        regressor = pickle.load(file)

    with open(sc_file, 'rb') as file:  
        sc = pickle.load(file)
        
    stk = yf.Ticker(S)
    df= stk.history(period="10y")
    df.iloc[:,0:5].to_csv(".//model_data//"+S+".csv")
    df = pd.read_csv(".//model_data//"+S+".csv",index_col="Date",parse_dates=True)
    
    df3= df['Open']
    df3= df3[-20:]
    plt.figure(figsize=(20,5))
    plt.plot(df3)
    plt.title(S+" Last 30 Days Actual Graph")
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
     
    plt.savefig(".//model_data//"+S+'_actual_graph.png')
    
    train_days= 60
    df= df[['Open']]
    df= df[-train_days:]
    
    testing_df= sc.transform(df)
    X_test = []
    X_test.append(testing_df[-train_days:])
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
    
    predicted_stock_price = regressor.predict(X_test)
    predicted_stock_price = sc.inverse_transform(predicted_stock_price)
    
    import datetime
    
    x=datetime.datetime.now() + datetime.timedelta(days=7)
    today_date= x.strftime("%x")
    
    df_write= pd.read_csv(".//model_data//"+S+"_predicted.csv")
    df_write.head()
    
    temp = pd.DataFrame([[today_date, predicted_stock_price[0][0]]], columns=['Date', 'Predicted Price'])
    if(len(df_write)==0 or df_write.iloc[-1,0]!=today_date):
        df_write= df_write.append(temp, ignore_index= True);
        
    df_write.to_csv(".//model_data//"+S+"_predicted.csv", index=False)
    
    print(S+" Done !!")
```
